```
df = spark.read.option("header",True).parquet("/mnt/datalake/int/jbr/forehearth/furnace_265/furnace_265_dcs.parquet")
```

```
display(df.take(10))
```

```
pri_f265_furnace_params:
  drop_cols: ['dcs_265_Timestamp', 'dcs_265_Timestamp_dt', 'dcs_265_etag', 'dcs_265_PartitionKey',
              'dcs_265_DateTime_dt']
  time_stamp_col: 'dcs_265_DateTime_timestamp'
```

```
from pyspark.sql import functions as f
```

```
hourly_furnace_df = (
        df.drop('dcs_265_Timestamp', 'dcs_265_Timestamp_dt', 'dcs_265_etag', 'dcs_265_PartitionKey',
              'dcs_265_DateTime_dt')
        .withColumn(
            "Timestamp",
            f.date_trunc("hour", f.col('dcs_265_DateTime_timestamp').cast("timestamp")),
        )
        .drop('dcs_265_DateTime_timestamp')
        .filter(
            (f.to_date("Timestamp") < "2020-03-24")
            | (f.to_date("Timestamp") > "2020-04-15")
            | (f.to_date("Timestamp") >= "2019-10-01")
        )
    )
```

```
display(hourly_furnace_df.take(10))
```

```
df2 = hourly_furnace_df.orderBy(f.col("Timestamp")).groupBy("Timestamp").agg(*([f.mean(f.col(i)).alias(i)
                    for i in hourly_furnace_df.columns
                    if ("Timestamp" or "RowKey") not in i
                ]
            )
        ).dropna().orderBy(f.col("Timestamp"))
```

```
display(df2.take(10))
```
