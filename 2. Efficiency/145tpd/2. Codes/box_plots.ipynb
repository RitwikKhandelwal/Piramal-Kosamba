```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
seaborn.set(style='whitegrid')
```

```python
df = pd.read_csv("Flat_File_Forehearth_Level_DPR_Eff_J20.csv")
```

```python
df.describe()["DPR Eff"]
```

```python
df["Type"] = ""*len(df)
```

```python
c1 = df.index[(df["DPR Eff"]>=0) & (df["DPR Eff"]<=54)]
c2 = df.index[(df["DPR Eff"]>=54) & (df["DPR Eff"]<=75)]
c3 = df.index[(df["DPR Eff"] >= 75) & (df["DPR Eff"] <= 85)]
c4 = df.index[(df["DPR Eff"] >= 85) & (df["DPR Eff"] <= 99)]

c1, c2, c3, c4
```

```python
df.loc[c1, "Type"] = "[0, 54]"
df.loc[c2, "Type"] = "[54, 75]"
df.loc[c3, "Type"] = "[75, 85]"
df.loc[c4, "Type"] = "[85, 99]"
```

```python
df1 = df.rename(columns = {"Type":"DPR_Eff_Metric"})
```

```python
li.append("DPR_Eff_Metric")
```

```python
for column in df1.columns:
    if(column[:10] == "DCS_160TPD" and column != "DCS_160TPD_PTHPR15AI72"):
        df1 = df1.rename(columns={column: "_".join(column.split("_")[2:-2])})
df1
```

```python
li = df1.columns.tolist()
li = li[:-1]
```

```python
li
```

```python
li
```

```python
for elem in li:
    if(elem=="Name" or elem=="Date" or elem=="FGCode"):
        continue
    seaborn.boxplot(x="DPR_Eff_Metric",
                y=elem,
                data=df1, order=["[0, 54]", "[54, 75]", "[75, 85]", "[85, 99]"])
    plt.savefig("Box Plots/Forehearth Level/J20/DPR Efficiency/{}.png".format(elem))
    plt.clf()
```
