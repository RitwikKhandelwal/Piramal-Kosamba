```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import glob as glob
from datetime import datetime
```

```python
onlyfiles = [f for f in listdir(
    '145tpd_1yr_Furnace_data/') if isfile(join('145tpd_1yr_Furnace_data', f))]
```

```python
onlyfiles
```

```python
only = []
```

```python
onlyfiles
for files in onlyfiles:
    f = ".".join(files.split(".")[:-1])
    only.append(f)
```

```python
only
```

```python
df_final = pd.DataFrame()
```

```python
for elem in only:
    df = pd.read_excel("145tpd_1yr_Furnace_data/{}.xls".format((elem)))
    df["Date Time"] = df['Date Time'].astype('datetime64[s]')
    df_final = pd.concat([df_final, df])
    # df2.to_csv("Month Wise Efficiency/{}.csv".format(elem), index=False)

    


```

```python
df_final
```

```python
# df_final = df_final.reset_index()
# df_final
# df_final = df.drop('index', axis=1)
# df_final = df.drop('level_0', axis=1)
df_final = df_final.loc[:, ~df_final.columns.str.contains('^level_0')]
df_final = df_final.loc[:, ~df_final.columns.str.contains('^index')]
```

```python
df_final
```

```python
df_final["Date Time"] = df_final['Date Time'].dt.round('30min')
```

```python
df_final = df_final.groupby("Date Time").mean().reset_index()
```
