```python
df = pd.read_csv('b5_flatfile_apr_may2022.csv')
```

```python
cols = df.columns.to_list()
```

```python
for i in range(3):
    cols.pop(0)
cols
```

```python
c1 = []
for i in range(2,59):
    c1.append(i)
c2 = [0,1,59,60]
```

```python
df["Type"] = ""*len(df)
df.loc[c1, "Type"] = "Good DPR"
df.loc[c2, "Type"] = "Bad DPR"
```

```python
for elem in cols:
    sns.boxplot(x="Type",
                y=elem,
                data=df, order=["Good DPR","Bad DPR"])
    plt.savefig('plots/anomaly_analysis/'+elem)
    plt.show()
```
