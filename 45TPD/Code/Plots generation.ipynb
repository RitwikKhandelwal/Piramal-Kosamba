# Box Plots

## B4

```python
df4 = pd.read_csv('b4_flatfile_apr_may.csv')
df4 = df4.iloc[:,1:]
df4.columns
```

```python
df4.describe()["DPR_Eff"]
```

```python
df4["Type"] = ""*len(df4)
c1 = df4.index[(df4["DPR_Eff"]>=0) & (df4["DPR_Eff"]<=73.48)]
c2 = df4.index[(df4["DPR_Eff"]>73.48) & (df4["DPR_Eff"]<=76.1)]
c3 = df4.index[(df4["DPR_Eff"]>76.1) & (df4["DPR_Eff"] <=82.2)]
c4 = df4.index[(df4["DPR_Eff"]> 82.2) & (df4["DPR_Eff"] <= 99)]
c1, c2, c3, c4
```

```python
df4.loc[c1, "Type"] = "[0, 73.48]"
df4.loc[c2, "Type"] = "[73.48, 76.1]"
df4.loc[c3, "Type"] = "[76.1, 82.2]"
df4.loc[c4, "Type"] = "[82.2, 99]"
```

```python
df41 = df4.rename(columns = {"Type":"DPR_Eff_Metric"})
```

```python
li = ['TC_MB_2_2_PV','TC_MC_2_2_PV','THAFCL_PV','TC_23FT_PV','CaO']
for elem in li:
    sns.boxplot(x="DPR_Eff_Metric",
                y=elem,
                data=df41, order=["[0, 73.48]", "[73.48, 76.1]", "[76.1, 82.2]", "[82.2, 99]"])
    plt.savefig('plots/box/b4/'+elem)
    plt.show()
```
